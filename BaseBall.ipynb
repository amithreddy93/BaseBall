```python
# Trying Kmeans for the first time to see the results only
from sklearn.cluster import KMeans
from sklearn import metrics

y = df.drop(['Wins'],axis=1)
score_dict = {}
for i in range(2,18):
    km = KMeans(n_clusters=i,random_state=1)
    l = km.fit_predict(y)
    s_s = metrics.silhouette_score(y,l)
    score_dict[i] =[s_s]

score_dict
```

```python
import numpy as np
kmeans_model = KMeans(n_clusters=3,max_iter=300,random_state=1)
dist = kmeans_model.fit_transform(y)

labels = kmeans_model.labels_
colormap = np.array(['r', 'g', 'b'])
plt.scatter(dist[:,0],dist[:,1],c=colormap[labels])
plt.title('Kmeans Clusters')
plt.show()
```

# Modeling Our Data

```python
df = df.drop(['labels'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(['Wins'],axis=1)
y=df['Wins']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 42)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report

LRmodel = LinearRegression()
LRmodel.fit(X_train,y_train)
predictions = LRmodel.predict(X_test)
test_score = r2_score(y_test,predictions)
```

```python

from sklearn.metrics import mean_absolute_error,mean_squared_error
print("The accuracy of testing dataset ",test_score*100)
print('MAE for testing set: {}'.format(mean_absolute_error(predictions,y_test)))
print('MSE for testing set: {}'.format(mean_squared_error(predictions,y_test)))
print('RMSE for testing set: {}'.format(np.sqrt(mean_squared_error(predictions,y_test))))
```

```python
# Compare against actual values
results = pd.DataFrame({'Actual':y_test,'Predicted':predictions})
results
```

```python
# Normalize the data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
multiple_linear_reg = LinearRegression(fit_intercept = False)
multiple_linear_reg.fit(X_train,y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_features = PolynomialFeatures(degree=5)
X_train_poly = polynomial_features.fit_transform(X_train)
X_test_poly = polynomial_features.fit_transform(X_test)

polynomial_reg = LinearRegression(fit_intercept = False)
polynomial_reg.fit(X_train_poly,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree_reg = DecisionTreeRegressor(max_depth = 5, random_state=13)
decision_tree_reg.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_reg = RandomForestRegressor(n_estimators = 400, max_depth = 5, random_state=13)
random_forest_reg.fit(X_train,y_train)
```

```python
from sklearn.svm import SVR

support_vector_reg = SVR(gamma='auto',kernel = 'linear',C=1)
support_vector_reg.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Evaluating Multiple Linear Regression Model

```python
# Predictions with training dataset
y_pred_MLR_train = multiple_linear_reg.predict(X_train)

# Predictions with testing dataset
y_pred_MLR_test = multiple_linear_reg.predict(X_test)

# Training accuracy for the model
accuracy_MLR_train = r2_score(y_train, y_pred_MLR_train)
print("Training accuracy for Multiple Linear Regression model: ", accuracy_MLR_train)

# Testing accuracy for the model
accuracy_MLR_test = r2_score(y_test,y_pred_MLR_test)
print("Testing accuracy for Multiple Linear Regression model: ", accuracy_MLR_test)

# RMSE for training data
RMSE_MLR_train = sqrt(mean_squared_error(y_train,y_pred_MLR_train))
print("RMSE for training data:", RMSE_MLR_train)

# RMSE for testing data
RMSE_MLR_test = sqrt(mean_squared_error(y_test,y_pred_MLR_test))
print("RMSE for testing data:", RMSE_MLR_test)

# Predictions with 10-Fold Cross Validation
y_pred_cv_MLR = cross_val_predict(multiple_linear_reg, X, y, cv=10)

# Accuracy after 10 - FOld Cross Validation
accuracy_cv_MLR = r2_score(y,y_pred_cv_MLR)
print("Accuracy for 10-Fold Cross Predicted MLR model: ", accuracy_cv_MLR)
```
